#Process Data for Feature Selection

```python
df['Gender'] = df['Gender'].replace(['M', 'F'], ['1', '2'])
df['Education_Level'] = df['Education_Level'].replace(['College', 'Doctorate', 'Graduate', 'High School', 'Post-Graduate', 'Uneducated', 'Unknown'], ['1', '2', '3', '4', '5', '6', '7'])
df['Marital_Status'] = df['Marital_Status'].replace(['Divorced', 'Married', 'Single', 'Unknown'], ['1', '2', '3', '4'])
df['Income_Category'] = df['Income_Category'].replace(['$120K +', '$40K - $60K', '$60K - $80K', '$80K - $120K', 'Less than $40K', 'Unknown'], ['1', '2', '3', '4', '5', '6'])
df['Card_Category'] = df['Card_Category'].replace(['Blue', 'Gold', 'Platinum', 'Silver'], ['1', '2', '3', '4'])
y = df["Attrition_Flag"]
X = df.drop("Attrition_Flag", axis =1)
```

#Feature Selection (Variance Threshold)

```python
from sklearn.feature_selection import VarianceThreshold
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

vt = VarianceThreshold(.5)
X_train = vt.fit_transform(X_train)
X_test = vt.transform(X_test)
```
